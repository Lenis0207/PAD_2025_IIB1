## 7. ¿Cuál es el precio promedio?

```python
# Precio promedio
precio_promedio = df_csv["price"].mean()

print("💰 Precio promedio:", precio_promedio)
```

## 10. Utiliza idxmax() para encontrar el índice del vino con el precio más alto y luego utiliza loc para obtener toda la información de ese vino específico.

```python
vinos_california = df_csv[df_csv["province"] == "California"]

indice_max_california = vinos_california["price"].idxmax()

vino_california_mas_caro = df_csv.loc[indice_max_california]

print("🍷 Vino más caro de California:\n")
print(vino_california_mas_caro)
```

## 11. ¿Cuáles son los tipos de uva más comunes en California?

```python
vinos_california = df_csv[df_csv["province"] == "California"]

tipos_uva_california = vinos_california["variety"].value_counts().head(44508)

print("🍇 Tipos de uva más comunes en California:\n")
print(tipos_uva_california)
```

## 12.  ¿Cuáles son los 10 tipos de uva más comunes en California?

```python
vinos_california = df_csv[df_csv["province"] == "California"]

top10_uvas_california = vinos_california["variety"].value_counts().head(10)

print("🍷 Top 10 variedades de uva en California:\n")
print(top10_uvas_california)
```
