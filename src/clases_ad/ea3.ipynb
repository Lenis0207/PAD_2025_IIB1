##  1. Crea un DataFrame frutas

```python
import pandas as pd

# Crear DataFrame
frutas = pd.DataFrame({
    "Granadilla": [20],
    "Tomates": [50]
})

print(frutas)
```

## 2. Crea un DataFrame ventas_frutas

```python
ventas_frutas = pd.DataFrame({
    "Granadilla": [20, 49],
    "Tomates": [50, 100]
}, index=["ventas 2021", "ventas 2022"])

print(ventas_frutas)
```

## 3. Crea una variable utensilios con una Serie

```python

utensilios = pd.Series(
    ["3 unidades", "2 unidades", "4 unidades", "5 unidades"],
    index=["Cuchara", "Tenedor", "Cuchillo", "Plato"],
    name="Cocina"
)

print(utensilios)
```
