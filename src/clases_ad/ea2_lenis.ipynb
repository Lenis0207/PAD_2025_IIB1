# 4. Matriz 4x4 donde cada elemento es i + j y su inversa

```python
matriz = np.fromfunction(lambda i, j: i * j + 1, (4, 4), dtype=int)
print("Matriz 4x4:\n", matriz)
```

```python
det = np.linalg.det(matriz)
print("\nDeterminante:", det)
```

```python
if det != 0:
    matriz_inversa = np.linalg.inv(matriz)
    print("\nInversa de la matriz:\n", matriz_inversa)
else:
    print("⚠️ La matriz es singular, no tiene inversa.")
```

```python
while True:
    matriz = np.random.randint(1, 100, (4, 4))  # valores aleatorios entre 1 y 101
    det = np.linalg.det(matriz)
    if det != 0:  # condición para que sea invertible
        break

print("Matriz invertible:\n", matriz)
print("\nDeterminante:", det)

matriz_inversa = np.linalg.inv(matriz)
print("\nInversa de la matriz:\n", matriz_inversa)
```

# 12. Dispersión con y = sin(x) + ruido vs sin(x) real

```python
# x= Nombre Estrato y= Unidades Cobradas
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y_real = np.sin(x)
ruido = np.random.normal(0, 0.2, size=x.shape)
y_ruido = y_real + ruido

plt.figure(figsize=(6, 4))
plt.scatter(x, y_ruido, s=15, alpha=0.6, label=r"$y = \sin(x) + \ ruido$")
plt.plot(x, y_real, color="red", linewidth=2, label=r"$y = \sin(x)$")
plt.title("Seno con ruido vs función real")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.show()
```

# 13. Gráfico de contorno usando np.meshgrid

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X) + np.sin(Y)

plt.figure(figsize=(6, 4))
contours = plt.contour(X, Y, Z, levels=20, cmap="viridis")
plt.clabel(contours, inline=True, fontsize=8)
plt.title(r"Contorno de $z = \cos(x) + \sin(y)$")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

# 14. Dispersión con densidad como color de los puntos

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
xy = np.vstack([x, y])
densidad = gaussian_kde(xy)(xy)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, c=densidad, s=20, cmap="plasma")
plt.title("Dispersión con densidad de puntos")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.colorbar(label="Densidad")
plt.show()
```

# 15. Gráfico de contorno lleno (filled contour)

```python
plt.figure(figsize=(6, 4))
plt.contourf(X, Y, Z, levels=20, cmap="viridis")
plt.colorbar(label="Valor de Z")
plt.title(r"Contorno lleno de $z = \cos(x) + \sin(y)$")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

# 17. Histograma con una distribución normal

```python
data = np.random.randn(1000)  # 1000 números de distribución normal estándar
plt.figure(figsize=(6,4))
plt.hist(data, bins=30, alpha=0.7, color="blue", edgecolor="black")
plt.title("Histograma de distribución normal")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

# 18. Dos distribuciones normales diferentes en el mismo histograma

```python
data1 = np.random.normal(loc=0, scale=1, size=1000)   # media 0, desviación 1
data2 = np.random.normal(loc=3, scale=1.5, size=1000) # media 3, desviación 1.5
plt.figure(figsize=(6,4))
plt.hist(data1, bins=30, alpha=0.6, label="N(0,1)", edgecolor="black")
plt.hist(data2, bins=30, alpha=0.6, label="N(3,1.5)", edgecolor="black")
plt.title("Dos distribuciones normales")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

# 19. Probar diferentes valores de bins

```python
plt.figure(figsize=(12,4))
for i, bins in enumerate([10, 30, 50], 1):
    plt.subplot(1,3,i)
    plt.hist(data, bins=bins, alpha=0.7, color="green", edgecolor="black")
    plt.title(f"bins = {bins}")
plt.tight_layout()
plt.show()
```

# 20. Línea vertical con la media

```python
media = np.mean(data)
plt.figure(figsize=(6,4))
plt.hist(data, bins=30, alpha=0.7, color="purple", edgecolor="black")
plt.axvline(media, color="red", linestyle="dashed", linewidth=2, label=f"Media = {media:.2f}")
plt.title("Histograma con línea de la media")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

# 21. Histogramas superpuestos (colores y transparencias)

```python
plt.figure(figsize=(6,4))
plt.hist(data1, bins=30, alpha=0.5, label="N(0,1)", color="blue", edgecolor="black")
plt.hist(data2, bins=30, alpha=0.5, label="N(3,1.5)", color="orange", edgecolor="black")
plt.title("Histogramas superpuestos")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```
